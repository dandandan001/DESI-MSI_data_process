```python
import pandas as pd
import numpy as np

# Read the dataset into csv file
# The file should contains all the features and intensities
# python counts index from 0
df = pd.read_csv("path/to/your/file.csv", header=[0], keep_default_na=False, na_values=[''])
df
```

```python
# adjust the number of columns based on the csv data file
# separate the pixel_id and coordinates from features(means m/z)
coor_columns = df.iloc[:, :3]
coor_columns
```

```python
# adjust the number of columns based on the csv data file
# separate the pixel_id and coordinates from features(means m/z)
peak_columns = df.iloc[:, 3:1003]
peak_columns
```

```python
# Each pixel contains a full mass spectra.
# Total Ion Current (TIC) normalization: normalize each peak (m/z) in each pixel based on the TIC of that pixel.
# Total Ion Current (TIC) of a pixel = row sum of all feature columns
# drop the rows that are only zeros
row_sums = peak_columns.sum(axis=1)
zero_rows = row_sums[row_sums == 0].index
peak_columns = peak_columns.drop(zero_rows)
# each inetnstiy times 10e5 to make it a larger number, easier to read, wont' change the comparison
TIC_norm = 10e5*peak_columns.div(peak_columns.sum(axis=1), axis=0)
TIC_norm
```

```python
# drop zero rows if exists
coor_columns = coor_columns.drop(zero_rows)
coor_columns
```

```python
# put the pixel_id and coordinates back with TIC results as a new dataframe
# save the results
TIC_result = pd.concat([coor_columns, TIC_norm], axis=1)
TIC_result.to_csv("path/to/your/output/folder/result.csv", index=None)
TIC_result
```
