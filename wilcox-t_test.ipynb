```r
# This script is modified based on Zongyuan Liu's script.
# Please cite https://github.com/zyliu-OU/McCall-Lab/tree/main/03172021

# laod your dataset file
# print out the first few row and dimensions of data
a<-read.csv("path/to/your/file.csv")
head(a)
dim(a)
```

```r
# sunset the data groups for calculation
# Here is to calculate between naive center and infected center groups. changed it according to your data.
features_pos<-droplevels(subset(a, a$group==c("naive_C")|a$group==c("infected_C")))
```

```r
# features starts from column 8 to column 830. Adjust the index based on your own data.
# subset only to those without a sum of zero
noMeta<-features_pos[,8:830] 
head(noMeta)
not_zero<-which(colSums(noMeta) != 0)
head(not_zero)
nonzero <- noMeta[, not_zero] # all the non-zero columns
nonzero
#add back metadata
nonzero_wMeta<-cbind(features_pos[,1:7], nonzero)
head(nonzero_wMeta)
dim(features_pos)
dim(noMeta)
dim(nonzero)
dim(nonzero_wMeta)
```

```r
pval<-c()#create an empty vector with value pval

# make sure to adjust here based on your feature columns index
for (i in 8:ncol(nonzero_wMeta)) 
pval<-c(pval,wilcox.test(nonzero_wMeta[,i]~nonzero_wMeta$group)$p.value) #assign to pval the output of the Mann-Whitney-Wilcoxon test
   
head(pval) #to check

# calculate adjusted p-val and print out first few rows
pAdj <- p.adjust(pval, method = c("fdr"), n=length(pval)) 
head(pAdj) 

```

```r
length(pval) # print out the length of pval for check
```

```r
length(pAdj) # print out the length of pAdj for check
```

```r
# make sure to adjust here based on your feature columns index
avg<- apply(nonzero_wMeta[,8:ncol(nonzero_wMeta)], 2, function(x) tapply(x, nonzero_wMeta$group, function(x) mean(x))) 

dim(avg) #to check

avg[,1:3]#to check

fch <- apply(avg, 2, function(x) x[1]/x[2]) #calculate fold change
# print out the results to see the fold change and check which group is x1 
head(fch) 
is(fch)
```

```r
# print out the fold change
fch
```

```r
#make dataframe from pval and fch and save the results as csv file
df <- data.frame(fch,pval, pAdj)
write.csv(df, "path/to/your/result_file.csv")
```
