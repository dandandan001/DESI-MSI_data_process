```python
import pandas as pd
import numpy as np
```

```python
# load the TIC normalized results here
# python index starts from 0
df = pd.read_csv("path/to/your/file.csv", header=[0], keep_default_na=False, na_values=[''])
df
```

```python
# load the file containing the pixels ids only belong to tissue region
# it can be get from Waters MSI segmentation software
file = pd.read_csv("path/to/tissue_only_pixels.csv")
WT_list = file.scan_num.to_list()
WT_list
```

```python
# Create a mask where True indicates the value in the first column is not in the list
mask = ~df.iloc[:, 0].isin(WT_list)

# mask all regions not belong to tissue
# Change the values in the rest of the columns based on the mask
df.loc[mask, df.columns[3:]] = 0  # Replace with 0 or any value you want
```
